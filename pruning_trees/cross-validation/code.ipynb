Splitting the data into features and target labels

```python
x_train=np.array(df.iloc[:,1:12])
y_train=np.array(df.iloc[:,12])
```

Encoding all the features which needs to be encoded for traing the model

```python
for i in range(x_train.shape[0]):
  if x_train[i][0]=='Male':
    x_train[i][0]=0
  if x_train[i][0]=='Female':
     x_train[i][0]=1
  if x_train[i][1]=='Yes':
     x_train[i][1]=1
  if x_train[i][1]=='No':
     x_train[i][1]=0
  if x_train[i][2]=='0':
     x_train[i][2]=1
  if x_train[i][2]=='1':
     x_train[i][2]=1
  if x_train[i][2]=='2':
     x_train[i][2]=2
  if x_train[i][2]=='3+':
     x_train[i][2]=3
  if x_train[i][3]=='Graduate':
     x_train[i][3]=0
  if x_train[i][3]=='Not Graduate':
     x_train[i][3]=1
  if x_train[i][4]=='Yes':
     x_train[i][4]=1
  if x_train[i][4]=='No':
     x_train[i][4]=0
  if x_train[i][10]=='Urban':
     x_train[i][10]=0
  if x_train[i][10]=='Semiurban':
     x_train[i][10]=1
  if x_train[i][10]=='Rural':
     x_train[i][10]=2
```

One Hot Encoding the target labels

```python
y_train = np.array([1 if label == 'Y' else 0 for label in y_train])
y_train
```

Normalising the continous features present in training data

```python
selected_feature_indices = [5, 6, 7, 8]
scaler = StandardScaler()
x_train[:, selected_feature_indices] = np.array(scaler.fit_transform(x_train[:, selected_feature_indices]))
x_train = x_train.astype(np.float16)
x_train = np.nan_to_num(x_train, nan=0.0)
```

```python
x_train[0]
```



*   Initialising the logistic regression model from sklearn with 'saga' solver and no regularising penalty with a maximum iterations of 5000.
*   initialising all the lists and necessary variables for 5 fold  cross validation strategy
*   Implemented Cross validation by splitting data at random indices and calculated the required metrics
*  finally calcualted the mean precison,mean accuracy and mean recall after cross validation









```python
model = LogisticRegression(solver='saga', penalty=None, max_iter=5000)

n_folds = 5
validation_percentage = 0.2
m = x_train.shape[0]
prec_score_list=[]
acc_score_list=[]
rec_score_list=[]
num_validation_samples = int(m * validation_percentage)


indices = np.arange(m)
np.random.shuffle(indices)
fold_size = m // n_folds
folds = [indices[i:i+fold_size] for i in range(0, m, fold_size)]


for fold_number in range(n_folds):
    validation_indices = folds[fold_number]
    training_indices = np.concatenate([f for i, f in enumerate(folds) if i != fold_number])

    X_train, Y_train = x_train[training_indices], y_train[training_indices]
    X_valid, Y_valid = x_train[validation_indices], y_train[validation_indices]

    model.fit(X_train, Y_train)
    Y_pred = model.predict(X_valid)
    prec_score = precision_score(Y_valid, Y_pred)
    acc_score = accuracy_score(Y_valid, Y_pred)
    rec_score = recall_score(Y_valid, Y_pred)
    prec_score_list.append(prec_score)
    acc_score_list.append(acc_score)
    rec_score_list.append(rec_score)
    print("fold:", fold_number + 1, "prec_score:", prec_score, "acc_score:", acc_score, "rec_score:", rec_score)

prec_score = sum(prec_score_list) / n_folds
acc_score = sum(acc_score_list) / n_folds
rec_score = sum(rec_score_list) / n_folds
print("mean precision:", prec_score)
print("mean acc_score:", acc_score)
print("mean rec_score:", rec_score)
```
